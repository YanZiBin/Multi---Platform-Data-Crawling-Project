```python
# 导入requests库用于发送HTTP请求
import requests
# 导入html解析器，用于解析网页内容
from lxml import html

# 设置目标小说网站的URL地址
url="https://www.biqgg.cc/book/10439/"
# 发送GET请求获取网页内容，并将二进制响应内容解码为UTF-8字符串
src=requests.get(url).content.decode('utf-8')
# 使用lxml的HTML解析器将字符串解析为可查询的HTML元素树
element=html.fromstring(src)
print("学号：222226205227")
```

```python
# 导入必要的库
import time  # 导入time模块用于计时
import pymongo  # 导入pymongo模块用于MongoDB数据库操作
import re  # 导入re模块用于正则表达式匹配

# 连接MongoDB数据库
client = pymongo.MongoClient("mongodb://localhost:27017/")  # 创建MongoDB客户端连接
db = client["我的天赋是复活"]  # 选择数据库
collection = db["单线程"]  # 选择集合

def process_chapter(chapter_info):
    """处理单个章节的函数"""
    # 获取章节信息
    title = chapter_info['title']  # 获取章节标题
    full_url = chapter_info['url']  # 获取章节URL
    
    print(f"开始处理: {title}")
    
    # 添加请求头，模拟浏览器访问
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
    }
    
    # 发送HTTP请求获取章节内容
    response = requests.get(full_url, headers=headers)  # 发送GET请求
    content_text = response.content.decode('utf-8')  # 将响应内容解码为UTF-8字符串
    
    # 使用正则表达式提取正文内容
    content = re.findall(r'<div id="chaptercontent".*?>(.*?)请收藏本站', content_text, re.S)
    
    # 构建要存入MongoDB的数据文档
    chapter_data = {
        "标题": title,
        "内容": content,
        "链接": full_url,
    }
    
    # 将数据插入MongoDB数据库
    collection.insert_one(chapter_data)
    print(f"已保存: {title}")
    
    return True

# 记录爬虫开始时间
start_time = time.time()

# 设置小说网站的基础URL
base_url = "https://www.biqgg.cc"

# 使用xpath获取所有章节链接
chapters = element.xpath('//div[@class="listmain"]/dl/dd[not(@class="more pc_none")]/a | //span[@class="dd_hide"]/dd/a')
print(f"总共找到 {len(chapters)} 个章节，将爬取前20章")
print("学号：222226205227")

# 获取前20章的链接信息并存储到列表中
chapters_info = []
for chapter in chapters[:20]:  # 只取前20章
    chapters_info.append({
        'title': chapter.text,  # 获取章节标题
        'url': base_url + chapter.get('href')  # 拼接完整的章节URL
    })

print("开始爬取...")

# 使用单线程方式逐个处理每个章节
for chapter_info in chapters_info:
    process_chapter(chapter_info)

# 计算爬虫总用时
total_time = time.time() - start_time

# 输出爬取结果统计信息
print("\n全部保存完成！")
print(f"总用时: {total_time:.2f} 秒")
```
