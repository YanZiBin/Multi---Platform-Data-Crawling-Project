```python
# 导入所需的库
from selenium import webdriver  # 用于浏览器自动化
from selenium.webdriver.common.by import By  # 用于定位元素
from selenium.webdriver.common.keys import Keys  # 用于模拟键盘操作
from selenium.webdriver.chrome.options import Options  # 用于设置Chrome浏览器选项
import time  # 用于添加延时
import re  # 用于正则表达式操作
from pymongo import MongoClient  # 用于MongoDB数据库操作

# 配置Chrome浏览器选项，绕过反爬虫检测
chrome_options = Options()
chrome_options.add_experimental_option("excludeSwitches", ["enable-automation"])  # 禁用自动化提示
chrome_options.add_experimental_option('useAutomationExtension', False)  # 禁用自动化扩展
chrome_options.add_argument('--disable-blink-features=AutomationControlled')  # 禁用自动化控制特征

# 初始化Chrome浏览器
driver = webdriver.Chrome(options=chrome_options)
# 注入JavaScript代码以绕过webdriver检测
driver.execute_cdp_cmd('Page.addScriptToEvaluateOnNewDocument', {
    'source': 'Object.defineProperty(navigator, "webdriver", {get: () => undefined})'
})

# 连接到本地MongoDB数据库
client = MongoClient('mongodb://localhost:27017/')
db = client['知乎']  # 选择数据库
collection = db['武汉有哪些好大学？']  # 选择集合

# 开始知乎登录流程
driver.get("https://www.zhihu.com")  # 打开知乎首页
time.sleep(1)  # 等待页面加载

# 点击密码登录选项
password_login = driver.find_element(By.XPATH, "//div[contains(text(),'密码登录')]")
password_login.click()
time.sleep(1)

# 输入用户名
username_input = driver.find_element(By.NAME, "username")
username_input.send_keys("15337161956")
time.sleep(1)

# 输入密码
password_input = driver.find_element(By.NAME, "password")
password_input.send_keys("YanZiBin2004619")
time.sleep(1)

# 提交登录表单
password_input.send_keys(Keys.RETURN)

# 等待用户手动完成验证码验证
input("请完成验证码验证后按回车继续...")
print('学号：222226205227')
print("登录成功！")

# 导航到目标问题页面
driver.get("https://www.zhihu.com/question/313278823")
time.sleep(2)  # 等待页面加载

# 提示用户手动滚动页面以加载更多回答
input("\n请手动滚动页面加载想要的回答数量，完成后按回车继续...")

# 初始化回答列表并获取所有已加载的回答元素
answers = []
answer_elements = driver.find_elements(By.CLASS_NAME, "ContentItem")

print(f"\n开始解析 {len(answer_elements)} 个回答...")

# 遍历每个回答元素并提取信息
for answer in answer_elements:
    # 尝试获取用户名，如果是匿名回答则标记为"匿名用户"
    user_link = answer.find_elements(By.CSS_SELECTOR, ".AuthorInfo-name .UserLink-link")
    if user_link:
        username = user_link[0].text
    else:
        username = "匿名用户"
    
    # 获取回答内容
    content = answer.find_element(By.CLASS_NAME, "RichText").text
    
    # 将用户名和回答内容添加到答案列表中
    answers.append({
        'username': username,
        'content': content
    })

# 如果成功获取到回答，则保存到数据库
if answers:
    # 将所有回答批量插入MongoDB数据库
    collection.insert_many(answers)
    print(f"\n数据已保存到MongoDB数据库")
    print(f"共保存了 {len(answers)} 个回答")

# 程序结束前等待用户确认
input("\n按回车键退出...")
# 清理资源
driver.quit()  # 关闭浏览器
client.close()  # 关闭数据库连接
```
